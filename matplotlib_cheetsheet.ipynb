# 概要
matplotlibでのグラフ描画について記載する。

```python
import numpy as np
import pandas as pd

# matplotlib系インポート
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
# 三角関数のsinとcosを表示する(最低限なグラフ描画例)
x = np.linspace(-np.pi, np.pi)

# figの初期化
fig = plt.figure(figsize=(10,6)) # figsizeで出力時の大きさを調整できる
# Figure内にAxesを追加()
# subplot(1行目,1列目,何番目)
# →分割したい行列数を先に、指定し、それは何番目なのかを指定するイメージ
ax1 = fig.add_subplot(111)
ax1.plot(x, np.cos(x))
ax1.plot(x, np.sin(x))
ax1.xaxis.set_major_locator(ticker.MultipleLocator(1)) # x軸の間隔を指定する
plt.show()
```
